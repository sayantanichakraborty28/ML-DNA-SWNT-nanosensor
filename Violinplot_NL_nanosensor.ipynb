```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Read the Excel sheet into a DataFrame
df = pd.read_excel('reg-newmols-en-model5.xlsx', engine='openpyxl')

# Create a violin plot
plt.figure(figsize=(12, 6))
sns.violinplot(data=df, inner="quartile", palette="viridis", linewidth=1.5, saturation=0.8)
sns.violinplot(data=df, inner="quartile", palette="coolwarm", linewidth=1.5, saturation=0.8)
sns.violinplot(data=df, inner="quartile", palette="viridis", linewidth=1.5, saturation=0.8)
sns.stripplot(data=df, color="black", size=4, jitter=True)

# Add a red dotted line at y-axis value 0.2 (our threshold below which signifies low response)
plt.axhline(y=0.2, color='red', linestyle='--', linewidth=3)


# Set y-axis range and ticks
#plt.ylim(0, 2)
#plt.yticks([0, 0.5, 1, 1.5, 2], fontsize=12)

# Calculate and print the median for each fingerprint
#medians = df.median()
#for i, median_value in enumerate(medians):
#    print(f"Median for Fingerprint {i + 1}: {median_value}")

# Add labels and title
#plt.xlabel('Fingerprint', fontsize=18)
plt.xlabel('Molecule Name', fontsize=18)
plt.ylabel(r'$\frac{\Delta F}{F}$', fontsize=20)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.title('Violin Plot of Optical Responses for Different Models for 5th best r2 score', fontsize=20)
plt.tight_layout()
# Show the plot
plt.savefig('violinpl-reg-newmols-en-model5.png', dpi=1200)
plt.show()
```
